```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# loading dataset 
df = pd.read_csv("world_population.csv")
```

```python
# cleaning data 
df.dropna(inplace=True)
df.rename(columns={"Country/Territory": "Country"}, inplace=True)
```

```python
# Basic stats
print(df.describe())
print(f"Total rows: {len(df)}")
```

```python
# Top 10 countries by population
top10 = df.nlargest(10, "2022 Population")
print(top10)
```

```python
# top 10 most populated countries(2022)
plt.figure(figsize=(10,6))
plt.bar(top10["Country"], top10["2022 Population"], color="skyblue")
plt.xticks(rotation=45)
plt.title("Top 10 Most Populated Countries (2022)")
plt.ylabel("Population")
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Population Share (Pie Chart)
plt.figure(figsize=(8, 8))
plt.pie(top10["2022 Population"], labels=top10["Country"], autopct='%1.1f%%', startangle=140)
plt.title("Population Share of Top 10 Countries (2022)")
plt.show()
```

```python
# Distribution of Countries by Population (Histogram)
plt.figure(figsize=(10, 6))
plt.hist(df["2022 Population"]/1e6, bins=30, color="green", edgecolor="black")
plt.title("Distribution of Countries by Population (in Millions)")
plt.xlabel("Population (Millions)")
plt.ylabel("Number of Countries")
plt.tight_layout()
plt.show()
```

```python
# Global Population Trend Over Time (Line Chart)
years = [col for col in df.columns if "Population" in col]
global_pop = df[years].sum()

plt.figure(figsize=(10, 6))
plt.plot(years, global_pop / 1e9, marker='o', linestyle='-', color='purple')
plt.title("Global Population Trend Over Time")
plt.ylabel("Population (Billions)")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Top 10 Countries by Growth (Barh chart)
top_growth = df.nlargest(10, "Total_Growth")
plt.figure(figsize=(10, 6))
plt.barh(top_growth["Country"], top_growth["Total_Growth"], color="orange")
plt.title("Top 10 Countries by Population Growth (1970-2022)")
plt.xlabel("Growth (%)")
plt.tight_layout()
plt.show()
```

```python
df["Total_Growth"] = ((df["2022 Population"] - df["1970 Population"]) / df["1970 Population"]) * 100
print(df.sort_values("Total_Growth", ascending=False).head())
```
