# loading a image in numpy and changing its color to dark

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# loading the image 
try:
    logo = np.load('numpy-logo.npy')

    #Displaying the loaded img
    plt.figure(figsize=(10,5))
    plt.subplot(121)
    plt.imshow(logo)
    plt.title("Numpy logo")
    plt.grid(False)

    
except FileNotFoundError:
    print("numpy logo file not found")   
```

```python
  # dark logo logic

try:
    logo = np.load('numpy-logo.npy')
    dark_logo = 1 - logo

    plt.subplot(122)
    plt.imshow(dark_logo)
    plt.title("Numpy Dark logo")
    plt.grid(False)

except FileNotFoundError:
    print("numpy logo file not found")  
```
